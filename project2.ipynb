```python
plt.figure(figsize=(2,4))
ax=sns.countplot(data=entertainer2,x="Gender (traditional)")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(20,4))
ax=sns.countplot(data=entertainer2,x="Year of Death")
ax.bar_label(ax.containers[0])
plt.show
```

```python
gb=entertainer2.groupby("Entertainer").agg({"Year of First Oscar/Grammy/Emmy":"mean","Year of Last Major Work (arguable)":"mean"})
print(gb)
```

```python
plt.figure(figsize=(50,100))
sns.heatmap(gb,annot=True)
plt.show()
```
